# To Analysis the Data's of the types of Problem in the Heart.

```
from IPython.core.interactiveshell import import_item
import os
import numpy as np
import pandas as pd
import seaborn as sns

```

```
df = pd.read_csv('heart.csv')
```

```
df.head()
```

# Now we'll plot a graph for Gender(sex) vs Count

```
sns.displot(df['Sex'], kde = True, color = 'black')
```

# Now we'll plot a graph for Age vs Count

```
sns.displot(df['Age'], kde = True, color = 'red')
```

# Now we'll plot a graph for Resting BP

```
sns.displot(x = df['RestingBP'], kde = True, color = 'Green')
```

```
sns.displot(x = df['Cholesterol'], kde = True, color = 'Green')
```

```
sns.displot(x = df['RestingECG'], kde = True, color = 'Green')
```

```
sns.displot(x = df['HeartDisease'], kde = True, color = 'Red')
```

# Now we'll analyze using pie chart to find the comparison between the ratios.

```
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f', figsize = (5,5))
```

```
df.groupby('RestingBP').size().plot(kind = 'pie', autopct = '%.1f', figsize = (15,30))
```

```
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f', figsize = (15,5))
```

```
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f', figsize = (5,5))
```

```
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f', figsize = (5,5))
```

# Now we're going to get the variation for the same ratios.

```
sns.violinplot(x = df['Age'], figsize = (5,5), color = 'Red')
```

# Compare the heart's problem vs the Sex

```
sns.violinplot(y = df['Sex'], x = df['Age'])
```

```
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])
```

```
sns.violinplot(y = df['Age'], x = df['HeartDisease'])

```

```
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'])
```
